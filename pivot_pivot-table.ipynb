The pivot function is used to create a new derived table out of a given one. Pivot takes 3 arguements with the following names: index, columns, and values. As a value for each of these parameters you need to specify a column name in the original table. Then the pivot function will create a new table, whose row and column indices are the unique values of the respective parameters. The cell values of the new table are taken from column given as the values parameter.

index : string or object, optional
--------
    Column to use to make new frame's index. If None, uses existing index.    
columns : string or object
--------
    Column to use to make new frame's columns.    
values : string, object or a list of the previous, optional
--------
    Column(s) to use for populating new frame's values. If not
    specified, all remaining columns will be used and the result will have hierarchically indexed columns.
Returns
-------
DataFrame
    Returns reshaped DataFrame.    

```python
df.pivot(index="date",columns="city")
```

```python
df.pivot(index="date",columns="city",values="humidity")
```

As above data has duplicate records hense we can not use pivot as used in below command we will get below error.

---------

```python
df.pivot(index="date",columns="city")
```

# This problem can be overcome by using "pivot_table"

```python
df.pivot_table(index="date",columns="city")
```

Signature:
----------
df.pivot_table(
    values=None,
    index=None,
    columns=None,
    aggfunc='mean',
    fill_value=None,
    margins=False,
    dropna=True,
    margins_name='All',
)

By default 'mean' aggregate function is provided

```python
df.pivot_table(index="date",columns="city",margins=True)
```
